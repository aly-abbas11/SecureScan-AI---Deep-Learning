# SecureScan AI - EDA Notebook (Week 1)

Exploratory Data Analysis for BigVul, DiverseVul, and FormAI datasets.

## Dataset Statistics
- BigVul: 188K samples (94% safe)
- DiverseVul: 319K samples (94% safe)
- FormAI: 246K samples (82% vulnerable)

## Preprocessing
- Clean code (remove comments)
- Balance data (3:1 ratio)
- Train/val/test split (70/15/15)
